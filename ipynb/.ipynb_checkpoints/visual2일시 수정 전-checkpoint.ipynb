# 호흡기질환 - 평균기온

```python

for i in [2016,2017,2018]:
    df_weather = pd.read_csv(f'C:/app/TeamProject/mod_weather_Data/기상관측{i}real_Fixed.csv', encoding = 'cp949')
    if i in [2016]:
        df_medi = pd.read_csv(f'C:/app/TeamProject/mod_medi_Data/NHIS_{i}_J_Cases.csv', encoding = 'cp949')
    else :
        df_medi = pd.read_csv(f'C:/app/TeamProject/mod_medi_Data/NHIS_{i}_J_Cases.csv', encoding = 'UTF-8')
        pass

    df_medi['요양개시일자'] =df_medi['요양개시일자'].astype('str')
    df_medi['일시(월별)'] = df_medi['요양개시일자'].apply(lambda x:x[:6])
    df_medi_count = df_medi.groupby('일시(월별)').count()['주상병코드']
    
    df_weather['일시'] =df_weather['일시'].astype('str')
    df_weather['일시(월별)'] = df_weather['일시'].apply(lambda x:x[:6])
    df_weather_mean = df_weather.groupby('일시(월별)')['평균기온(°C)'].mean()
    df_medi_count
    
    f, ax = plt.subplots(figsize = (20, 7))
    ax =sns.barplot(x = df_medi_count.index,
                     y = '주상병코드',
                    data = df_medi_count.to_frame(),
                   color = 'skyblue')
    ax2 = ax.twinx()
    ax2 = sns.lineplot(x = df_weather_mean.index,
                      y = '평균기온(°C)',
                      data = df_weather_mean.to_frame(),
                      color = 'red')

#     ax1.set_ylim(0, 50000)
    ax2.set_ylim(-5, 30)
#     ax1.set_xlabel(f'{i}', size = 10)
    ax.set_ylabel('월별 진료건수(명)', size = 10)
    ax2.set_ylabel('월 평균기온(°C)', size = 10)
    plt.grid(True)
#     plt.axis([1950, 2010, 30, 90])
    plt.title(f'{i}년 월평균기온과 월별 진료건수')
    plt.ylabel('Number of Cases', fontsize=12)
    plt.savefig(f'{i}년 월평균기온과 월별 진료건수', dpi = 700)
    plt.show()
    pass
```

## 호흡기질환(시도별) - 평균기온

```python
for i in [2016, 2017, 2018]:
    df_weather = pd.read_csv(f'C:/app/TeamProject/mod_weather_Data/기상관측{i}real_Fixed.csv', encoding = 'cp949')
    if i in [2016]:
        df_medi = pd.read_csv(f'C:/app/TeamProject/mod_medi_Data/NHIS_{i}_J_Cases.csv', encoding = 'cp949')
    else :
        df_medi = pd.read_csv(f'C:/app/TeamProject/mod_medi_Data/NHIS_{i}_J_Cases.csv', encoding = 'UTF-8')
        pass
    df_medi['요양개시일자']=df_medi['요양개시일자'].astype('str')
    df_weather['일시']=df_weather['일시'].astype('str')
    df_medi['일시(월별)'] = df_medi['요양개시일자'].apply(lambda x:x[:6])
    df_weather['일시(월별)'] = df_weather['일시'].apply(lambda x:x[:6])
    
    
    
    for j in sidocode_list :
        medi_sido = df_medi[df_medi['시도코드'].isin([f'{j}'])]
        weather_sido = df_weather[df_weather['시도코드'].isin([f'{j}'])]
        
        medi_sido_group = medi_sido.groupby('일시(월별)').count()['주상병코드']
        medi_sido_group['진료건수'] = medi_sido_group['주상병코드']
        weather_sido_group = weather_sido.groupby('일시(월별)')['평균기온(°C)'].mean()
        
        f, ax = plt.subplots(figsize = (20, 7))
        ax =sns.barplot(x = medi_sido_group.index,
                         y = '진료건수',
                        data = medi_sido_group.to_frame(),
                       color = 'skyblue' )
        ax2 = ax.twinx()
        ax2 = sns.lineplot(x = weather_sido_group.index,
                          y = '평균기온(°C)',
                          data = weather_sido_group.to_frame(),
                          color = 'red')

#         ax.set_ylim(0, 50000)
        ax2.set_ylim(-5, 30)
#         ax2.set_ylabel('평균기온(°C)', fontsize=12)
        
#         ax2.set_ylabel('평균기온(°C)', fontsize=12)
        z = sido_code[j]
        ax.set_ylabel('진료건수')
        plt.grid(True)
        plt.title(f'{i}년 {z} 평균기온과 진료건수')
#         plt.ylabel('평균기온(°C)', fontsize=12)
            
        
        
        plt.savefig(f'{i}년 {z} 평균기온과 진료건수', dpi = 700)
        plt.show()
        pass
    pass
```

# 호흡기질환 - 미세먼지

```python
for i in [2016,2017,2018]:
    df_dust = pd.read_csv('C:/app/TeamProject/mod_dust_Data/fine_dust_data.csv', encoding = 'cp949')
    if i in [2016]:
        df_medi = pd.read_csv(f'C:/app/TeamProject/mod_medi_Data/NHIS_{i}_J_Cases.csv', encoding = 'cp949')
    else :
        df_medi = pd.read_csv(f'C:/app/TeamProject/mod_medi_Data/NHIS_{i}_J_Cases.csv', encoding = 'UTF-8')
        pass
    df_medi['요양개시일자']=df_medi['요양개시일자'].astype('str')
    df_dust['일시']=df_dust['일시'].astype('str')
    
    df_dust['일시(월별)'] = df_dust['일시'].apply(lambda x:x[:6])
    df_medi['일시(월별)'] = df_medi['요양개시일자'].apply(lambda x:x[:6])
    df_dust = df_dust[df_dust['일시'].str.startswith(f'{i}')]
    df_medi_count = df_medi.groupby('일시(월별)').count()['주상병코드']
    df_medi_count
    df_dust_mean = df_dust.groupby('일시(월별)').mean()
    
    f, ax = plt.subplots(figsize = (15, 7))
    ax =sns.barplot(x = df_medi_count.index,
                     y = '주상병코드',
                    data = df_medi_count.to_frame(),
                   color = 'skyblue')
    ax2 = ax.twinx()
    
    ax2 =sns.lineplot(x = df_dust_mean.index,
                      y = 'PM10', 
                     label = '미세먼지(PM10)',
                     data = df_dust_mean)
    sns.lineplot(x = df_dust_mean.index,
                 y = 'PM25', 
                 label = '초미세먼지(PM25)',
                 data = df_dust_mean)
    
    ax.set_xlabel(f'{i}', size = 10)
    ax.set_ylabel('진료건수(명)', size = 10)
    ax2.set_ylabel('미세먼지농도(PM10)', size = 10)
    plt.grid(True)
    plt.title(f'{i}년 월평균 미세먼지와 월별 진료건수')
    plt.savefig(f'{i}년 월평균 미세먼지와 월별 진료건수', dpi = 700)
    plt.show()
    pass
```
